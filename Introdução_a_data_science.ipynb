# Melhorando a visualizaÃ§Ã£o

```python
total_lingua = tmdb["original_language"].value_counts()
total_geral = total_lingua.sum()
total_ingles = total_lingua.loc["en"]
total_resto = total_geral - total_ingles
print(total_geral, total_ingles, total_resto)
```

```python
dados = {
    "lingua" : ["ingles", "outros"],
    "total" : [total_ingles, total_resto]
}

dados = pd.DataFrame(dados)
dados
```

```python
sns.barplot(data=dados, x="lingua", y="total")
```

```python
dados.plot(kind="pie", y="total", labels=dados["lingua"])
```

```python
total_filmes_outras_linguas = tmdb.query("original_language != 'en'")["original_language"].value_counts()
total_filmes_outras_linguas.head()
```

```python
sns.countplot(data=tmdb.query("original_language != 'en' "), x = "original_language")
```

```python
tmdb.query("original_language != 'en'")["original_language"].value_counts(normalize=True)
```

```python
# grafico plotado na ordem

plt.figure(figsize=(16,8))
sns.countplot(data=tmdb.query("original_language != 'en' "),
              order =  total_filmes_outras_linguas.index,
              hue = "original_language",
              hue_order = total_filmes_outras_linguas.index,
              palette = "mako",
              stat = "percent",
              x = "original_language")
plt.title("Quantidade de filmes por lÃ­ngua exceto em inglÃªs")
```

```python
notas_toyStory = notas.query("IdFilme == 1")["Nota"]
notas_jumanji = notas.query("IdFilme == 2")["Nota"]
notas_toyStory.head()
```

```python
media_toyStory = notas_toyStory.mean()
media_jumanji = notas_jumanji.mean()
print(media_toyStory, media_jumanji)
```

```python
# mediana
mediana_toyStory = notas_toyStory.median()
mediana_jumanji = notas_jumanji.median()
print(mediana_toyStory, mediana_jumanji)
```

```python
# mostrando diferenÃ§a de notas dispersas
import numpy as np

filme1 = [2.5] * 10 + [3.5] * 10
filme2 = [5] * 10 + [1] * 10


media_filme1 = np.mean(filme1)
mediana_filme1 = np.median(filme1)

media_filme2 = np.mean(filme2)
mediana_filme2 = np.median(filme2)

print("Filme 1:")
print("MÃ©dia: ", media_filme1)
print("Mediana: ", mediana_filme1)

print("\nFilme 2:")
print("MÃ©dia: ", media_filme2)
print("Mediana: ", mediana_filme2)
```

```python
plt.hist(filme1)
plt.hist(filme2)
```

```python
plt.boxplot([filme1, filme2])
```

```python
filme0 = [3] * 20 # espalha 0
np.mean(filme0)
```

```python
# desvio padrÃ£o
np.std(filme0), np.std(filme1), np.std(filme2)
```
